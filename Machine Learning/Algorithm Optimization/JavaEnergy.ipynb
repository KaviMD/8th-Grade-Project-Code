```python
# Pandas for managing datasets
import pandas as pd

# Numpy and Random for math
import numpy as np
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)

data = pd.read_csv('Data/Windows.csv')
energy_data = pd.read_csv('Data/EnergyOutput.csv')
data = clean_dataset(data)
#data = data.query('windowSize==63')
print len(data['energyg1'])
print len(energy_data['gx'])
data['energyg1'] = energy_data['gx'].values
data['energyg2'] = energy_data['gy'].values
data['energyg3'] = energy_data['gz'].values
data['energya1'] = energy_data['ax'].values
data['energya2'] = energy_data['ay'].values
data['energya3'] = energy_data['az'].values

data.to_csv('Data/JavaEnergy.csv', sep=',', index=False)
data.head()
```

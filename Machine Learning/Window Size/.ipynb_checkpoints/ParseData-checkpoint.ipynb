```python
# Pandas for managing datasets
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting and styling
import seaborn as sns
```

```python
data = pd.read_csv('Data\\Results.csv')
```

```python
data['params']=data['param1'].map(str)+data['param2'].map(str)+data['param3'].map(str)+data['param4'].map(str)
```

```python
data_bestParams = data.groupby(['Algorithm','params'])['accuracy'].mean().reset_index()
```

```python
data_best = data[((data['Algorithm']=='GNB') & (data['params']=='NoneNoneNoneNone')) | ((data['Algorithm']=='LR') & (data['params']=='l10.010.0001liblinear')) | ((data['Algorithm']=='RFC') & (data['params']=='1005NoneNone'))]
data_best = data_best.groupby(['Algorithm','windowSize'])['accuracy'].mean().reset_index()
```

```python
data_best['windowSizeSeconds'] = (data_best['windowSize']/3)*200
```

```python
ax = sns.pointplot(x="windowSizeSeconds", y="accuracy", hue='Algorithm',data=data_best)
```
